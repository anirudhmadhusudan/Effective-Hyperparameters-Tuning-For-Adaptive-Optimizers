```python
# Larger CNN for the MNIST Dataset
import numpy
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.layers import Flatten
from keras.layers.convolutional import Conv2D
from keras.layers.convolutional import MaxPooling2D
from keras.utils import np_utils
from keras import optimizers
from keras import backend as K
K.set_image_dim_ordering('th')
# fix random seed for reproducibility
seed = 7
numpy.random.seed(seed)
# load data
(X_train, y_train), (X_test, y_test) = mnist.load_data()
# reshape to be [samples][pixels][width][height]
X_train = X_train.reshape(X_train.shape[0], 1, 28, 28).astype('float32')
X_test = X_test.reshape(X_test.shape[0], 1, 28, 28).astype('float32')
# normalize inputs from 0-255 to 0-1
X_train = X_train / 255
X_test = X_test / 255
# one hot encode outputs
y_train = np_utils.to_categorical(y_train)
y_test = np_utils.to_categorical(y_test)
num_classes = y_test.shape[1]
```

```python
# define the larger model
def MNIST_AdaGrad(lrate,epsilonvalue=1e-7,decayvalue=0):
      
    model = Sequential()
    model.add(Conv2D(30, (5, 5), input_shape=(1, 28, 28), activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Conv2D(15, (3, 3), activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Dropout(0.2))
    model.add(Flatten())
    model.add(Dense(128, activation='relu'))
    model.add(Dense(50, activation='relu'))
    model.add(Dense(num_classes, activation='softmax'))
    adagrad = optimizers.Adam(lr=lrate, epsilon=epsilonvalue, decay=decayvalue)
    model.compile(loss='categorical_crossentropy', optimizer=adagrad, metrics=['accuracy'])
    return model
```

```python
lrates = [0.01,0.001,0.0001,0.00001]

for lr in lrates:
    model = MNIST_AdaGrad(lr)
    # Fit the model
    model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=15, batch_size=200)
    # Final evaluation of the model
    scores = model.evaluate(X_test, y_test, verbose=0)
    print("Large CNN Error: %.2f%%" % (100-scores[1]*100))
```

```python
def MNIST_RMSprop(lrate,rhovalue=0.9, epsilonvalue=1e-7,decayvalue=0):
      
    model = Sequential()
    model.add(Conv2D(30, (5, 5), input_shape=(1, 28, 28), activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Conv2D(15, (3, 3), activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Dropout(0.2))
    model.add(Flatten())
    model.add(Dense(128, activation='relu'))
    model.add(Dense(50, activation='relu'))
    model.add(Dense(num_classes, activation='softmax'))
    rmsprop = optimizers.RMSprop(lr=lrate,rho=rhovalue,epsilon=epsilonvalue,decay=decayvalue)
    model.compile(loss='categorical_crossentropy', optimizer=rmsprop, metrics=['accuracy'])
    return model
```

```python
lrates = [0.01,0.001,0.0001,0.00001]

for lr in lrates:
    model = MNIST_RMSprop(lr)
    # Fit the model
    model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=15, batch_size=200)
    # Final evaluation of the model
    scores = model.evaluate(X_test, y_test, verbose=0)
    print("Large CNN Error: %.2f%%" % (100-scores[1]*100))
```

```python
def MNIST_Adam(lrate,beta1=0.9,beta2=0.99,epsilonvalue=1e-7,decayvalue=0):
      
    model = Sequential()
    model.add(Conv2D(30, (5, 5), input_shape=(1, 28, 28), activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Conv2D(15, (3, 3), activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Dropout(0.2))
    model.add(Flatten())
    model.add(Dense(128, activation='relu'))
    model.add(Dense(50, activation='relu'))
    model.add(Dense(num_classes, activation='softmax'))
    adam = optimizers.Adam(lr=lrate, beta_1=beta1, beta_2=beta2, epsilon=epsilonvalue, decay=decayvalue)
    model.compile(loss='categorical_crossentropy', optimizer=adam, metrics=['accuracy'])
    return model
```

```python
lrates = [0.01,0.001,0.0001,0.00001]

for lr in lrates:
    model = MNIST_Adam(lr)
    # Fit the model
    model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=15, batch_size=200)
    # Final evaluation of the model
    scores = model.evaluate(X_test, y_test, verbose=0)
    print("Large CNN Error: %.2f%%" % (100-scores[1]*100))
```

```python
beta1values = [0.8,0.9,0.95,0.975,0.98,0.99]
lr = 0.001

for beta1 in beta1values:
    model = MNIST_Adam(lr,beta1=beta1)
    # Fit the model
    model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=15, batch_size=200)
    # Final evaluation of the model
    scores = model.evaluate(X_test, y_test, verbose=0)
    print("For beta1 = ",beta1,", Large CNN Error: %.2f%%" % (100-scores[1]*100))
```

```python
beta1values = [0.975,0.98,0.99]
beta2values = beta2values = [0.98,0.995,0.999]
lr = 0.001

for beta1 in beta1values:
    for beta2 in beta2values:
        model = MNIST_Adam(lr,beta1=beta1,beta2=beta2)
        # Fit the model
        model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=20, batch_size=200)
        # Final evaluation of the model
        scores = model.evaluate(X_test, y_test, verbose=0)
        print("For beta1 = ",beta1," , For beta2",beta2,", Large CNN Error: %.2f%%" % (100-scores[1]*100))
```
